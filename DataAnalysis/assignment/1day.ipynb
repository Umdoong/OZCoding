```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris
```

```python
# 결측치가 있는 행을 모두 제거하세요.
iris.dropna(inplace=True)
iris
```

```python
# 중복된 행을 제거하세요.
iris.drop_duplicates(inplace=True)
iris
```

```python
# 'petal_length'가 보다 큰 행만 선택하세요.
filtered_iris = iris[iris['petal_length'] > 5]
filtered_iris.sort_values(by='petal_length', ascending=True)
```

```python
# 'species'에 따라 데이터를 그룹화하세요
grouped_iris = iris.groupby('species')
grouped_iris.describe()
```

```python
# 그룹화한 데이터에서 'sepal_length'의 평균을 계산하세요
grouped_iris['sepal_length'].mean()
```

```python
# 'petal_length'를 기준으로 데이터를 오름차순으로 정렬하세요.
iris.sort_values(by='petal_length', ascending=True)
```

```python
# 'petal_width'를 기준으로 데이터를 내림차순으로 정렬하세요
iris.sort_values(by='petal_width', ascending=False)
```

```python
# 'species'의 빈도수를 구하세요
iris['species'].value_counts()
```

```python
# 'species'값 중 'setosa'를 포함하는 행만 선택하세요.
iris[iris['species'].str.contains('setosa')]
```

```python
# 'species'값 중 'setosa'로 시작하는 행을 선택하세요
iris[iris['species'].str.startswith('setosa')]
```

```python
# 'species'값이 'virginica'로 끝나는 행을 선택하세요.
iris[iris['species'].str.endswith('virginica')]
```

```python
# 'species'의 'senosa'를 'SETOSA'로 대체하세요
iris['species'] = iris['species'].str.replace('setosa', 'SETOSA')
iris
```

```python
# iris 데이터셋에서 'species'열을 제거하세요.
iris.drop('species', axis=1, inplace=True)
iris
```

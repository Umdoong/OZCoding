[문제2]

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

컴퓨터는 실수를 부동 소수점 방식으로 표현합니다.  
부동 소수점 방식이란 말 그대로 소수점을 움직이는 방식입니다.  
부동 소수점은 10진수의 경우 m(가수)x10^n(지수)의 꼴로 나타내는데
예를 들어 100.4 라는 실수가 있을 때 부동 소수점 방식으로 1.004 x 10^2 또는 1004.0 x 10^-1로 표현할 수 있습니다.  
우리는 평소에 10진수를 사용하지만 오늘날 컴퓨터는 2진수의 부동 소수점(1.m x 2^n)으로 실수를 저장합니다.  
문제는 여기서 발생합니다.  
0.3의 경우 m x 3^의 꼴로 나타내고 싶다면 1 x 3^-1로 값이 딱 맞아 떨어지게 표현할 수 있습니다.  
하지만 10진수를 사용한 m x 10^n으로 표현한다면 0.333333...처럼 3이 무한히 반복되는 순환소수가 됩니다.
이와 같은 문제가 10진수와 2진수 사이에서도 나타납니다.  
10진수 0.1은 1x10^-1로 간단하게 표현할 수 있지만 1.m x 2^n으로 표현하면 무한하거나 무한에 가까운 소수가 됩니다.  
하지만 컴퓨터의 공간은 한정적이고 무한히 많은 소수를 저장할 수 없습니다.  
때문에 소수점을 중간에 생략하게 되는데 이로 인해 오차가 발생하게 됩니다.  
<br>

##### TMI  
파이썬은 실수 타입이 float 하나 밖에 없습니다.
Java에서는 float, double 타입이 있는데 float은 32비트르 이용한 부동 소수점 방식을 사용하고  
double은 64비트를 이용한 부동 소수점 방식을 사용합니다. 이름도 double인 이유가 float의 2배이기 때문입니다.
파이썬의 float은 64비트를 사용하기 때문에 double과 같은 저장 공간을 가지고 있습니다.  
그럼 메모리 효율을 위해 32비트를 써야할 때도 있을텐데 못쓰는 걸까요?
구글링해보니 array, struct 모듈 그리고 numpy라이브러리를 통해 32비트로 타입을 지정해줄 수 있는 방식이 있었습니다.  
<br>
2진수로 표현하는 부동 소수점 방식의 가수에는 왜 1이 들어갈까요?  
오늘날 컴퓨터는 IEEE 754 방식으로 실수를 저장합니다.  
32비트 => 부호(플래그에서 양수인지 음수인지 판단해주기 위함, 1bit), 지수(8bit), 가수(23bit)  
64비트 => 부호(0이면 음수 1이면 양수, 1bit), 지수(11bit), 가수(52bit)  
이런 방식으로요!!  
여기서 가수의 정수부에 1로 통일된 정규화한 수를 IEEE에서 지정해놓은 겁니다.
